```python
T_oi = [[1.0, 0.0, 0.0, 0.71],
         [0.0, 1.0, 0.0, -0.0165],
         [0.0, 0.0, 1.0, 0.42],
         [0.0, 0.0, 0.0, 1.0]]
```

```python
T_ci = [[0.005459969472053433, -0.9998896692056289, 0.013814418164449227, 0.0754590075458481],
         [0.0021750466690124703, -0.013802716671227067, -0.9999023723266596, -0.019547802789885316],
         [0.9999827288035288, 0.005489483432150222, 0.0020994442827294013, -0.010584573188241865],
         [0.0, 0.0, 0.0, 1.0]]
```

```python
T_ci_inv = np.linalg.inv(T_ci)
```

```python
T_oc = np.dot(T_oi, T_ci_inv)
T_oc
```

```python
R = T_oc[:3,:3]
R
```
